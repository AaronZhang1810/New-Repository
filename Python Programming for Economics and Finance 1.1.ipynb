```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = (10,6)

# Fixing random state for reproducibility
np.random.seed(19680801)

# Compute pie slices
N = 20
θ = np.linspace(0.0, 2 * np.pi, N, endpoint=False)
radii = 10 * np.random.rand(N)
width = np.pi / 4 * np.random.rand(N)
colors = plt.cm.viridis(radii / 10.)

ax = plt.subplot(111, projection='polar')
ax.bar(θ, radii, width=width, bottom=0.0, color=colors, alpha=0.5)

plt.show()
```

```python
r = 0.025         # interest rate
T = 50            # end date
b = np.empty(T+1) # an empty NumPy array, to store all b_t
b[0] = 10         # initial balance

for t in range(T):
    b[t+1] = (1 + r) * b[t]

plt.plot(b, label='bank balance')
plt.legend()
plt.show()
help(plt.plot)
help(plt.legend)
```

```python
%matplotlib inline 
import numpy as np
import matplotlib.pyplot as plt

T = 200
α= 0.9
ϵ = np.random.randn(T+1)
x = np.empty(T+1)
x[0] = 0

for t in range(T):
    x[t+1] = α*x[t] + ϵ[t+1]

plt.plot(x,label='correlated time series')
plt.legend()
plt.show()
```

```python
α_values = [0.0, 0.8, 0.98]
T = 200
x = np.empty(T+1)

for α in α_values:
    x[0] = 0
    for t in range(T):
        x[t+1] = α * x[t] + np.random.randn()
    plt.plot(x, label=f'$\\alpha = {α}$')

plt.legend()
plt.show()
```

```python
%matplotlib inline 
import numpy as np
import matplotlib.pyplot as plt

α = 0.9
T = 200
x = np.empty(T+1)
x[0] = 0

for t in range(T):
    if x[t]>=0:
        x[t+1]=α*x[t]+np.random.randn()
    else:
        x[t+1]=-α*x[t]+np.random.randn()
plt.plot(x, label=f'$\\alpha = {α}$')
plt.legend()
plt.show()
```

```python
α = 0.9
T = 200
x = np.empty(T+1)
x[0] = 0

for t in range(T):
    abs_x = - x[t] if x[t] < 0 else x[t] #if语句简便方法
    x[t+1] = α * abs_x + np.random.randn()

plt.plot(x)
plt.show()
```

```python
n = 100000

count = 0
for i in range(n):
    u, v = np.random.uniform(), np.random.uniform() #连续赋值
    d = np.sqrt((u - 0.5)**2 + (v - 0.5)**2) # **表示开方
    if d <= 0.5:
        count += 1

area_estimate = count / n

print(area_estimate * 4)  # dividing by radius**2
```

```python
import numpy as np

n = 100000
circle = 0

for i in range(n):
    x = np.random.uniform()
    y = np.random.uniform()
    d = np.sqrt((x-0.5)**2 + (y-0.5)**2)
    if d <= 0.5:
        circle += 1

print(4*circle/n)
```
