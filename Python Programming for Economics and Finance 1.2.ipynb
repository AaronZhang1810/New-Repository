```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (10,6)

def generate_data(n,type):
    ϵ = []
    for i in range(n):
        if type == "U" or "u":
            e = np.random.uniform(0,1)
        else:
            e = np.random.randn()
        ϵ.append(e)
    return ϵ

data = generate_data(100,'U')
plt.plot(data)
plt.show()
    
```
